```python
import os
import warnings
warnings.simplefilter("ignore")
import numpy as np 
import pandas as pd
```

```python
from skimage.io import imread,imshow
from skimage.transform import resize
from skimage.color import rgb2gray
```

```python
mads = os.listdir("/content/Mor")
```

```python
emile = os.listdir("/content/new")
```

```python
leto = os.listdir("/content/tabu")
```

```python
limit = 40
```

```python
mads_image=[None]*limit
j=0
for i in mads:
    if(j<limit):
        mads_image[j]=imread("/content/Mor/"+i)
        j+=1
    else:
        break
```

```python
emile_image=[None]*limit
j=0
for i in emile:
    if(j<limit):
        emile_image[j]=imread("/content/new/"+i)
        j+=1
    else:
        break
```

```python
leto_image=[None]*limit
j=0
for i in leto:
    if(j<limit):
        leto_image[j]=imread("/content/tabu/"+i)
        j+=1
    else:
        break
```

```python
imshow(leto_image[1])
```

```python
mads_gray=[None]*limit

j=0

for i in mads:

    if(j<limit):

        mads_gray[j]=rgb2gray(mads_image[j])

        j+=1

    else:

        break
```

```python
emile_gray=[None]*limit

j=0

for i in emile:

    if(j<limit):

        emile_gray[j]=rgb2gray(emile_image[j])

        j+=1

    else:

        break
```
